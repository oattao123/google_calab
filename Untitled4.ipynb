```
def createhtml(inputname,imgurl,oat):
  htmltag = '''
  <!DOCTYPE html>
  <html>
  <title>HTML Tutorial</title>
  <body>

  <h1>Hello my name is %s</h1>
  <p>This is my dog's picture</p>
  <img src="%s" alt="my image">
  <img src="%s" alt="my image">
  </body>
  </html>
  '''%(inputname,imgurl,oat)
  return htmltag
```

```
htmltag = createhtml('Oat','https://weloveanimals.in.th/wp-content/uploads/2020/02/202002-45.jpg','https://weloveanimals.in.th/wp-content/uploads/2020/02/202002-45.jpg')
```

```
a = display(ip.display.HTML(htmltag))
```
