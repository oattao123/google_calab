```
class car:
  def __init__(self,color,speed,name):
    self.color = color
    self.speed = speed
    self.position = 0
    self.name = name
  def changename(self,newname):
    self.name = newname
  def moveforward(self,hours):
    self.position = self.position + self.speed * hours
  def movebackward(self,hours):
    self.position = self.position - self.speed * hours
```
